<p style = "background-color:lightgray; font-size:20px; font-weight:bold">
    Task 11.    Remove outliers from dataset.
</p>

```python
# Calculate quartiles and IQR 
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
```

```python
# Filter rows 
mask = ~((df < (Q1- 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)
```

```python
df = df[mask]
```
