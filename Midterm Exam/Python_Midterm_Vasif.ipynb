### Task 3. Preparation

```python
# remove outliers
def outlier_bounds(col):
    Q1 = np.percentile(col,25)
    Q3 = np.percentile(col,75)
    IQR = Q3-Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR

    return lower_bound, upper_bound
```

```python
lower, upper = outlier_bounds(df['age'])
```

```python
df['age'] = np.clip(df['age'],a_min=lower, a_max=upper)
```

```python
# outliers
plt.figure(figsize=(10,7))
sns.boxplot(data=df, x='age')
```

### Task 4. Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('y',axis=1)
y = df['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=111, shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
```

```python
y_pred = logreg.predict(X_test)
y_prob = logreg.predict_proba(X_test)[:,1]
```

```python
print('Accuracy: ', accuracy_score(y_test,y_pred))
print('Precision: ', precision_score(y_test,y_pred))
print('Recall: ', recall_score(y_test,y_pred))
print('ROC&AUC: ', roc_auc_score(y_test,y_prob))
```

```python
## LGBM Classifier
```

```python
from lightgbm import LGBMClassifier
```

```python
lgbm_ = LGBMClassifier()
```

```python
lgbm_.fit(X_train,y_train)
```

```python
y_pred = lgbm_.predict(X_test)
y_prob = lgbm_.predict_proba(X_test)[:,1]
```

```python
print('Accuracy: ', accuracy_score(y_test,y_pred))
print('Precision: ', precision_score(y_test,y_pred))
print('Recall: ', recall_score(y_test,y_pred))
print('ROC&AUC: ', roc_auc_score(y_test,y_prob))
```

According to ROC&AUC and metrics LightGBM is better than Logistic Regression in this case.
