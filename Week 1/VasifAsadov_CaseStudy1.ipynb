Task 2.  Create a function called "addition" that can add two or more numbers till user enters 
zero and returns answer and count of input numbers in a list.

```python
def ask_inputs():
    numbers = []
    while True: 
        num = float(input("Enter number: "))
        if num==0:
            break 
        numbers.append(num)
    return numbers 
```

```python
# function 
def addition(*args): 
    return sum(args), len(args)

# call the function 
# numbers = ask_inputs()
addition_result, count_ = addition(*ask_inputs()) 
print("Sum: ", addition_result)
print("Count of arguments: ", count_)
```

Task 3. Create a function called "subtraction" that can subtract two or more numbers till user 
enters zero and returns answer and countof input numbers in a list. 

```python
# function 
def subtraction(*args): 
    if len(args) == 0: 
        return 0 

    result = args[0]
    for i in args[1:]:
        result-=i 
    return result, len(args)

# call the function 
subtraction_result, count_ = subtraction(*ask_inputs()) 
print("subtraction: ", subtraction_result)
print("Count of arguments: ", count_)
```

Task 4. Create a function called "multiplication" that can multiply two or more numbers till 
user enters zero and returns answer and count of input numbers in a list. 

```python
# function 
def multiplication(*args): 
    if len(args) == 0: 
        return 0 

    result = args[0]
    for i in args[1:]:
        result*=i 
    return result, len(args)

# call the function 
multiplication_result, count_ = multiplication(*ask_inputs()) 
print("Multiplication: ", multiplication_result)
print("Count of arguments: ", count_)
```

Task 5. Create a function called “average” that returns mean of “addition” function. 

```python
def average(*args): 
    sum_,count_ = addition(*args)
    return sum_/count_

print("Average: ", average(*ask_inputs()))
```

Task 6. Create a simple calculator that asks user to perform operations among created functions. 

```python
# ask input function 
def ask_inputs():
    numbers = []
    while True: 
        num = float(input("Enter number: "))
        if num==0:
            break 
        numbers.append(num)
    return numbers 
```

```python
# Calculator 
msg1 = input("Show operations: yes/no?")
if msg1.lower() == "yes" or msg1.lower() == "y": 
    msg2 = input("Choose one: Addition (1), Subtraction (2), Multiplication (3), Average (4)")
    if msg2 == "1" or msg2.lower() == "addition": 
        numbers = ask_inputs()
        sum_, count_ = addition(*numbers)
        print("Sum: ", sum_)
        print("Count: ", count_) 
    elif msg2 == "2" or msg2.lower() == "subtraction":
        numbers = ask_inputs()
        sub, count_ = subtraction(*numbers) 
        print("Subtraction: ", sub)
        print("Count: ", count_)
    elif msg2 == "3" or msg2.lower == "multiplication": 
        numbers = ask_inputs()
        mult, count_ = multiplication(*numbers)
        print("Multiplication: ", mult)
        print("Count: ", count_)
    else: 
        numbers = ask_inputs()
        avg = average(*numbers)
        print("Average: ", avg)
else: 
    print("You could not start operations! Restart again!")

```

Task 7. Write a Python program to find intersection of two given arrays using lambda.

```python
a = [1, 2, 3, 4,5,7,6,7,11,8]
b = [2,5,44,8,77,65,43,23]

intersection = list(filter(lambda x: x in b,a))
intersection 
```

Task 9. Write a Python program that stores numbers in a list which are divisible by nineteen or 
thirteen from a given list.

```python
divisible_by_19_13 = [19*13,19*13*888,19*13*2,19*13*54,19*13*877]
random.seed(11)
numbers = [random.randint(0,10000) for _ in range(30)] 
numbers.extend(divisible_by_19_13)
random.shuffle(numbers) 

numbers 
```

```python
result = list(filter(lambda x: x%13==0 and x%19==0, numbers))
result 
```
