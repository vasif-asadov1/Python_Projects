Task 1. Create a regex that can select names and ages from the given list and create a new dictionary with name as a key and age as a value: 

Nameage = "\  
Janice is 22 and Theon is 33 \ 
Gabriel is 44 and Joey is 21\ 
" 

```python
Nameage = """\  
Janice is 22 and Theon is 33 \ 
Gabriel is 44 and Joey is 21\ """
```

```python
# the criteria for the names is these: 
# 1. These are separate word blocks inside the text. So for separate word blocks
#    we should use \b...\b
# 1. must start with Capital letter [A-Z]
# 2. remaining letters of word must be lowercase [a-z]+ (at least 1 or more letters)
# based on conditions our pattern must be: \b[A-Z][a-z]+\b


# the conditions for characters being ages are:
# 1. must be separated block: \b..\b
# 2. Must consist of 1,2 or 3 digits. Because age can not consists of 4 digits.
# Resulted pattern: \b\d{1,3}\b
```

```python
names = re.findall(r"\b[A-Z][a-z]+\b", Nameage) 
ages = re.findall(r"\b\d{1,3}\b", Nameage)
NameAge = dict(zip(names,ages))
NameAge 
```

Task 2. Create a regex for email verification. All E-mail addresses should include:
-  3 to 10 lowercase or uppercase letters, numbers
-  An @ symbol  
-  2 to 10 lowercase and uppercase letters, numbers  
-  .
-  2 to 4 lowercase and uppercase letters

```python
# Conditions for being right email:
# 1. must be lowercase, uppercase letters or numbers: [a-zA-Z0-9]
# 2. 3 to 10 number limit: {3,10} 
# 3. @ symbol: @
# 4. 2 to 10 low. and upp. letters and numbers: [a-zA-Z0-9]{2,10}
# 5. must have "." dot: the dot character indicates everything until the new line. 
#   but here we should note it as a simple .(dot) sign. so we must use \ before it: \.
# 2 to 4 low. and upp. letters: [a-zA-Z]{2,4}

# Resulted pattern: [a-zA-Z0-9]{3,10}@[a-zA-Z0-9]{2,10}\.[a-zA-Z]{2,4}
```

```python
pattern = r"[a-zA-Z0-9]{3,10}@[a-zA-Z0-9]{2,10}\.[a-zA-Z]{2,4}"
text = "sadskf@.com"
def check_email(text):
    pattern = r"[a-zA-Z0-9]{3,10}@[a-zA-Z0-9]{2,10}\.[a-zA-Z]{2,4}" 
    if re.findall(pattern, text): 
        return True 
    else:
        return False 
    
if check_email(text):
    print("Email verified")
else:
    print("Invalid email")
```

Task 3. Create a regex for number verification. The order of numberâ€™s structure should be as    
following: 
1. ( 
2. 3 numbers 
3. ) 
4. Space 
5. numbers 
6. "-" 
7. numbers

```python
# Conditions: 
# 1. Starts with "(": ^\( (because () is notation in regex. 
#   so we should remove this notation using \ before it)
# 2. 3 numbers: \d{3} ({3}=exactly 3 digits))
# 3. ")" sign after numbers: \)
# 4. Space: \s 
# 5. numbers: \d+ (at least 1 or more digits)
# 6. "-" sign is simple: -
# numbers: \d+  

# Resulted pattern: ^\(\d{3}\)\s\d+-\d+
# Alternative pattern ^\([0-9]{3}\)\s[0-9]+-[0-9]+
```

```python
text = "(994) 012-13-32"

pattern = r"^\(\d{3}\)\s\d+-\d+"

if len(re.findall(pattern,text)) != 0:
    print("Valid phone number")
else: 
    print("Invalid phone number")

```
