<font color = "Blue">Task 4.  Follow the steps:

A) Define a structured data type ( name - S20, Surname - S20, age-i1, mark-f4) 

B) Create an array with the given information below in a data type created  
(Behram Abbasov 26 85, Yusif Abdullayev 22 92, Maryam Mecidova 19 88, Vagif Hesenzade 24 79) </font>

````python
# data types 
dtype = np.dtype([
    ('name', 'S20'), 
    ('surname', 'S20'), 
    ('age', 'i1'), 
    ('mark', 'f4')
])

# Create array 
people = np.array([
    ('Behram', 'Abbasov', 26, 85), 
    ('Yusif', 'Abdullayev', 22, 92), 
    ('Maryam', 'Mecidova', 19, 88), 
    ('Vagif', 'Hesenzade', 24, 79) 
], dtype=dtype)

people 
````

````python
arr = np.full((10,10), 99)
arr[1:9, 1:9] = 0
arr[1,1] = 1 
arr[len(arr)-2, len(arr)-2] = 1 
arr[2:-2, 2:-2] = 1
arr[3:-3, 3:-3] = 0 
arr[4:-4, 4:-4] = 1
arr 

````
